# 1. Digits model

## 1.1 Preprocessing

```python
from sklearn.preprocessing import OneHotEncoder

one_hot = OneHotEncoder()

shape = (20, 20)
X_train, X_test, y_train, y_test = train_test_split(X_digits, y_digits, test_size=0.15)
y_train_enc = one_hot.fit_transform(y_train.reshape(-1, 1)).toarray().astype(int)
y_test_enc = one_hot.transform(y_test.reshape(-1, 1)).toarray().astype(int)
```

## 1.2 Digits Neural Architecture and Evaluation

```python
cost = 'mse'
dnn_digits = DNN(n_iter=3000, eta=0.005, cost_function=cost, 
                 batch_size=500, algorithm='adam', regularization=None, 
                 cost_ratio=300, shuffle=True, print_cost=True, 
                 beta1=0.9, beta2=0.99, Lambda=0.05)

hidden_units = 60
hidden_layers = 8
output_units = len(np.unique(y_digits))
functions = {'sigmoid': sigmoid, 'tanh': tanh, 'relu': relu}
deriv_functions = {'sigmoid': deriv_sigmoid, 'tanh': deriv_tanh, 'relu': deriv_relu}

func = 'relu'

if func == 'relu':
    assert cost in ['mse', 'cross_entropy'], 'Change the cost function if you are using relu' 
    
elif func == 'sigmoid' or func == 'tanh':
    assert hidden_layers <= 3, f'With {func} you will suffer from vanishing gradient'
    
function, deriv = functions[func], deriv_functions[func]

# input layer
dnn_digits.add_layer(Layer(function, deriv, hidden_units), n_features=X_digits.shape[1])

# hidden layers
for _ in range(hidden_layers):
    dnn_digits.add_layer(Layer(function, deriv , hidden_units))
    
# output layer
dnn_digits.add_layer(Layer(softmax, deriv_softmax, output_units))

if True:
    dnn_digits.fit(X_train.T, y_train_enc.T)
    y_train_pred, y_train_prob = dnn_digits.predict(X_train.T, softmax=True)
    y_test_pred, y_test_prob = dnn_digits.predict(X_test.T, softmax=True)

    print('Train accuracy: ', np.round(accuracy_score(y_train_pred, y_train), decimals=3))
    print('Test accuracy: ', np.round(accuracy_score(y_test_pred, y_test), decimals=3))
```

# 3. Getting Equations

```python
letter_labels = np.array(['X', 'Y', 'Z'])

def predict_number(model, number):
    number = number.ravel().reshape(-1, 1).astype(bool)
    pred, prob = model.predict(number, softmax=True)
    return pred, prob

def predict_letter(model, letter):
    letter = letter.ravel().reshape(-1, 1).astype(bool)
    pred, prob = model.predict(letter, softmax=True)
    pred = letter_labels[pred]
    return pred, prob

def predict_algebra(symbol, threshold_area=200):
    n, labels, stats, _ = cv.connectedComponentsWithStats(symbol, 4, cv.CV_32S)
    n = n - 1
    if not n == 2:
        area = np.sum(symbol)
        if area < threshold_area:
            return '+'
        return '-'
    return '='
```

```python
equation = cv.imread('data/equation.png', 0)
imshow(equation)

block = equation_block(equation, show=True)
block = block[:, 20:]

block_blur = cv.GaussianBlur(block, (3, 3), 1, 1)
equations = identify_equations(block_blur, 
                               kernel=np.ones((2, 16)), 
                               show=True, 
                               iterations=6)

equations_info = {}
algebraic_symbols = []
number_symbols = []
letter_symbols = []

for n, equation in enumerate(equations):
    
    symbols, region = identify_characters(block, equation, kernel=np.ones((15, 3)))
    symbols = final_prep(symbols, region.copy(), show=False, shape=(20, 20))
    
    equations_info[n] = type_symbols = {'letters': [], 'numbers': [],
                                        'algebraic': [], 'right_side': []}
    n_blocks = len(symbols) // 3
        
    for i in range(0, len(symbols), 3):
        if len(symbols[i: i + 3]) == 3:
            
            number, letter, alg = symbols[i: i + 3]  
            
            if n == 2:
                algebraic_symbols.append(alg)
                number_symbols.append(number)
                letter_symbols.append(letter)
            
            imshow(number)
            imshow(letter)
            imshow(alg)
            
            equations_info[n]['letters'].append(predict_letter(dnn_letters, 
                                                               letter)[0])
            
            equations_info[n]['numbers'].append(predict_number(dnn_digits, 
                                                               number)[0])
            
            equations_info[n]['algebraic'].append(predict_algebra(alg))

        else:
            
            for number in symbols[i: i + 3]:
                equations_info[n]['right_side'].append(predict_number(dnn_digits, 
                                                                      number)[0])
                
                imshow(number)
    
    if True:
        print('Equation: ', n)
        print(" ")
        for key in type_symbols.keys():
            print(f'{key}', equations_info[n][key])
```

```python
real_equations = [['1X + 6Y + 5Z = 4'], ['3X + 8Y + 0Z = 3'], ['7X - 2Y - 9Z = 5']]

def write_equation(equation_info):
    keys = ['letters', 'numbers', 'algebraic', 'right_side']
    letters, numbers, algebra, right_side = [equation_info[key] for key in keys]
    
    output = ""
    for (l, n, alg) in zip(letters, numbers, algebra):
        output += str(n) + l + ' ' + alg + ' '
    for n in right_side:
        output += str(n)
        
    return output

for n, key in enumerate(equations_info.keys()):
    real_equation, = real_equations[n]
    predicted_equation = write_equation(equations_info[key])
    accuracy = (np.array(list(real_equation)) == np.array(list(predicted_equation))).sum()
    accuracy /= len(list(real_equation))
    print('Equation: ', n)
    print(" ")
    print('Real equation: ', real_equation)
    print('Predicted equation: ', predicted_equation)
    print(f'Accuracy of equation: ', accuracy)
    print(" ")
```
